```python
# imports
import yfinance as yf
import pandas as pd
```

```python
# Input the ticker symbols you want to take and save them into a data set this will be a list of dictionaries. 

tickers = ['TSLA','AAPL','MSFT']

# take ticker list, output csvs with info
def tick_hist(tickers):
    for t in tickers:
        df = pd.DataFrame(yf.Ticker(t).history(period='max'))
        df.columns = [i.lower() for i in df.columns]
        df.to_csv(f'./data/{t}_history.csv')
    return

tick_hist(tickers)
```

```python
def get_stock(ticker):
        stock = yf.Ticker(ticker).history()
        stock.columns = [i.lower() for i in stock.columns]
        stock = stock.pct_change()
        stock[['dividends','stock splits']] = stock[['dividends','stock splits']].fillna(0)
        stock.dropna(inplace=True)

        return stock
    
stock = get_stock(ticker)
type(stock)    
```
